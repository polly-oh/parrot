```python
# 8-1. [nan값, S, M, L]을 대상으로 20번 random sampling한 (5,4) 데이터프레임 df를 만드세요 
# (복원추출, probabilty값과 행index, column명은 재량)
```

```python
import numpy as np
df = pd.DataFrame(np.random.choice([np.NaN, 'S','M','L'],20,replace = True,).reshape(5,4),index = list('abcde'),columns = list('ABCD'));df
```

```python
# 8-2. level별 빈도값을 coulumn별로 산출하세요
# 조건: 행 index 순서는 S, M, L, nan 으로 출력하고, 값들을 int로 출력하세요
```

```python
A=df['A'].value_counts();B=df['B'].value_counts();C=df['C'].value_counts();D=df['D'].value_counts();
pd.concat([A,B,C,D],axis=1).reindex(index = ['S','M','L','nan']).fillna(0).astype(int)
```
